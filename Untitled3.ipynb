```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


y = np.random.rand(42)

ll = []
for i in range(2, 20):
    poly = PolynomialFeatures(degree=i)
    x_poly = poly.fit_transform(x)
    x_train, x_test, y_train, y_test = train_test_split(x_poly, y, test_size=0.3, random_state=40)
    l = LinearRegression()
    l.fit(x_train, y_train)
    y_pred = l.predict(x_test)
    value = mean_squared_error(y_test, y_pred)
    ll.append(value)

print(ll)
plt.plot(range(2, 20), ll)

plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score as acc_score  
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Desktop/Fish.csv")
df
```

```python
x = df.drop("Species", axis = 1)
y = df['Species']
scale = StandardScaler()
x_scale = scale.fit_transform(x)
x_scale
```

```python
k = range(1,16)
accuracy_scores = [] 
for i in k:
    knn = KNeighborsClassifier(n_neighbors=i)
    x_train, x_test, y_train, y_test = train_test_split(x_scale, y, test_size=0.25, random_state=21)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracy_scores.append(acc_score(y_test, y_pred))  
plt.plot(accuracy_scores)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score 
```

```python
x = df.drop("Species", axis = 1)
y = df['Species']
accuracy_scores = [] 
k = range(1,5)
for i in k:
    dt = DecisionTreeClassifier(criterion="entropy", max_depth = i)
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=21)
    dt.fit(x_train,y_train)
    y_pred = dt.predict(x_test)
    accuracy_scores.append(accuracy_score(y_test, y_pred))  
plt.plot(accuracy_scores,k)
plt.show()
```
