```python
df = pd.read_csv("Desktop/Churn.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df["TotalCharges"] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
```

```python
df.drop(columns =['customerID','Unnamed: 0'], inplace = True)
```

```python
df.columns
```

```python
category = ['gender', 'Partner', 'Dependents',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']
df = pd.get_dummies(df, columns = category, drop_first = True)
df
```

```python
x = df.drop('Churn', axis = 1)
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 32)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion='entropy', max_depth = 3)
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
# x columns has to be converted into numerical values , y accept labels, but conversion of x is necessary
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

feature_names = x_train.columns.tolist()  

plt.figure(figsize=(20,20))
tree.plot_tree(dt, filled=True, feature_names=feature_names)
plt.show()
```
