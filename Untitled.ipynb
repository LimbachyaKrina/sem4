```python
df = pd.read_csv("Desktop\CO2 Emissions.csv")
df
# as we are doing simple linear regression which means that there would be one x and one y
```

```python
# 1. Data Cleaning
df.isna().sum()
# drop null values, drop anomolyies, drop duplicates
```

```python
df.corr()
# have only numerical column
```

```python
x = df[['Fuel Consumption City (L/100 km)']]
y = df[['CO2 Emissions(g/km)']]
# double brackets for slicing
# if single then it gives series which means 1D but i want 2D data
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 3)
x_train
```

```python
from sklearn.linear_model import LinearRegression # this is a class
```

```python
lr = LinearRegression()
# this is a blank model
lr.fit(x_train,y_train)
# now this model is ready to predict the values
y_pred = lr.predict(x_test)
y_pred
```

```python
for i,j in zip(y_test['CO2 Emissions(g/km)'], y_pred):
    print(i,'-->',j)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# sum(act ^2 - pred^2)/ total no. of iteration
```

```python
mse = mean_squared_error(y_test, y_pred)
print(mse)
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
print(lr.coef_) # slope m
print(lr.intercept_) #c
```

```python
# to find the random state where r2  is maximum
max = 0
store = 0
for i in range(0,42):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = i)
    lr = LinearRegression()
    lr.fit(x_train,y_train)
    y_pred = lr.predict(x_test)
    r2 = r2_score(y_test, y_pred)
    if(r2 > max):
        max = r2
        store = i
print(store)
```
